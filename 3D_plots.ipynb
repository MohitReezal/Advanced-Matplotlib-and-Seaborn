```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#3D scatter plot
```

```python
batters=pd.read_csv('/content/batter(1).csv')
batters
```

```python

fig= plt.figure()
ax= plt.subplot(projection='3d')
ax.scatter3D(batters['runs'],batters['avg'], batters['strike_rate'])
ax.set_title('IPL batsman analysis')
ax.set_xlabel('Runs')
ax.set_ylabel('Avg')
ax.set_zlabel('Strike_rate')
```

```python
x=[0,1,5]
y=[0,10,13]
z=[0,13,20]
fig= plt.figure()
ax= plt.subplot(projection='3d')
ax.scatter3D(x,y,z)
ax.plot3D(x,y,z,color='red')
```
