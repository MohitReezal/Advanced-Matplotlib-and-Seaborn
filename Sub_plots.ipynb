```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#sub plot is a different way to plot graphs(specially multiple graphs)
```

```python
df=pd.read_csv('/content/batter(1).csv')
df.head()
```

```python
fig, ax= plt.subplots()
ax.scatter(df['avg'],df['strike_rate'])
ax.set_title('avg vs strike rate')
ax.set_xlabel('Avg')
ax.set_ylabel('Strike Rate')
```

```python
fig, ax= plt.subplots()
```

```python
fig,ax = plt.subplots(nrows=2,ncols=1,sharex= True)
ax[0].scatter(df['avg'],df['strike_rate'],color='green')
ax[1].scatter(df['avg'],df['runs'])
ax[0].set_title('Avg vs Strike rate')
ax[0].set_ylabel('Strike rate')
ax[1].set_title('Avg vs runs')
ax[1].set_ylabel('runs')
```

```python
fig,ax = plt.subplots(nrows=2,ncols=2,figsize=(10,8))
ax[0,0].scatter(df['avg'],df['strike_rate'],color='green')
ax[0,1].scatter(df['avg'],df['runs'], color= 'red')
ax[1,0].hist(df['avg'], color='violet')
ax[1,1].hist(df['runs'], color= 'violet')
```
